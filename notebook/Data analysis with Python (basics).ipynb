
## Types Of Data Analysis

### 1.) Descriptive Analytics or What happened in business :
“Describes Things”. It is the examination of data or content generated from events, usually manually performed, to answer the question “What happened?”. (Case studies,observations,surveys)e.g. Google Analytics

### 2.) Diagnostic Analytics or Why it happened in business:
examines data or content to answer the question “Why did it happen?”, a nd is characterized by techniques such as drill-down, data dscovery, data mining and correlations. e.g root cause analysis of fall of Nokia in India

### 3.) Predictive Analytics or what is likely to happen: 
It uses the discoveries of descriptive and diagnostic analytics for finding tendencies and predicting future trends by techniques such as regression analysis, forecasting, multivariate statistics, pattern matching, predictive modelling, and forecasting. e.g. Churn rate forecasting

### 4.) Prescriptive Analytics or what action needs to be taken:
The prescriptive model utilities an understanding of what has happened, why it has happened and a variety of “what-might-happen” analysis to help the user determine the best course of action to take. It s characterized by techniques such as graph analysis, simulation, complex event processing, neural networks, recommendation engines, machine learning. e.g. churn management

### 5.)Exploratory Data Analysis(EDA): 
It is an approach to analyzing data sets to summarize their main characteristics.

### 6.) Confirmatory Data Analysis(CDA): 
It focuses on confirming or falsifying existing hypotheses.( i.e addresses the truthfulness or falsehood of current theories

```python
import numpy as np


l = [i for i in range(2, 7)]
print(l)

l *= 2
print(l)

array1 = np.array(l)

print(array1)

print(array1 * 2)
```

```python
array1=np.array([(2,3,4,5),(1,0,3,6)])

print(array1.reshape(4,2))
print('\nTranspose\n')
print(array1.T)
```

```python
array1=np.array([(2,3,4,5),(1,0,3,6),(12,3,14,5),(11,-1,7,9), ])

# Stats

print('sum ', array1.sum())
print('max ', array1.max())
print('min ', array1.min())
print('std ', array1.std())
print('var ', array1.var())
print('median ', np.median(array1))
# print('sqrt ', np.sqrt(array1))
```

```python
# slicing
print(array1[1])
print(array1[1, 3])

# first two row with items till 0, 1, 2
print(array1[:2, 2:])

print(array1[:2, 2])
```

```python
# multiplication

array3=np.arange(10).reshape(5,2)   ##5*2=10,5*2 matrix/array
print(array3)

print(array3.T)
## becomes 2*5 matrix/array
## dot product
print(np.dot(array3.T,array3))
```

• In polls and surveys it can be used to measure the level of confidence of our results.

```python
dframe1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list('bcd'),index=['Ohio', 'Texas', 'Colorado'])
print(dframe1)
dframe2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list('bde'),index=['Utah', 'Ohio', 'Texas', 'Oregon'])
print(dframe2)
print(dframe1+dframe2)  ##union
print(dframe1-dframe2)  ## simple substraction
```
